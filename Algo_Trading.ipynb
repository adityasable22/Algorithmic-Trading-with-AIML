```python

# Import necessary libraries
import pandas as pd

# Load the Excel dataset
file_path = '/content/drive/MyDrive/btech project/algo trading dataset.xlsx'
df = pd.read_excel(file_path)

# Display the first few rows of the dataset
df.head()
```

```python
# Check for missing values
missing_values = df.isnull().sum()

# Display missing values
print("Missing values in each column:")
print(missing_values)
```

```python
# Drop rows with missing values
df_cleaned = df.dropna()

# Check the shape of the cleaned dataset
print(f"Shape of the dataset after dropping missing values: {df_cleaned.shape}")

# Preview the cleaned dataset
df_cleaned.head()
```

```python
# Drop rows where the 'profit/loss' column is equal to 0
df_cleaned = df[df['Profit / Loss'] != 0]

# Check the shape of the dataset after dropping these rows
print(f"Shape of the dataset after removing rows with 0 profit/loss: {df_cleaned.shape}")

# Preview the cleaned dataset
df_cleaned
```

```python
# Drop rows with missing values after removing 0 profit/loss rows
df_cleaned = df_cleaned.dropna()

# Check the shape of the dataset after dropping missing values
print(f"Shape of the dataset after dropping missing values: {df_cleaned.shape}")

# Preview the cleaned dataset
df_cleaned

```

```python
# Check the data types of each column
print("Data types before conversion:")
print(df_cleaned.dtypes)

# Convert necessary columns to numeric types (if not already)
# Replace 'RSI', 'MACD', 'CCI' with actual column names from your dataset
df_cleaned['RSI'] = pd.to_numeric(df_cleaned['RSI'], errors='coerce')
df_cleaned['MACD'] = pd.to_numeric(df_cleaned['MACD'], errors='coerce')
df_cleaned['CCI'] = pd.to_numeric(df_cleaned['CCI'], errors='coerce')
df_cleaned['Profit / Loss'] = pd.to_numeric(df_cleaned['Profit / Loss'], errors='coerce')

# Check the data types after conversion
print("\nData types after conversion:")
print(df_cleaned.dtypes)
```

```python
# Ensure the data is sorted by date (replace 'date_column' with the actual column name)
df_cleaned = df_cleaned.sort_values(by='Date')

# Set the date column as the index if necessary
df_cleaned.set_index('Date', inplace=True)
```

```python
df_cleaned
```

```python
# Define a function to categorize Profit/Loss based on Close prices
def categorize_profit_loss(row):
    # Calculate the percentage of profit/loss relative to the close price
    profit_loss_percentage = row['Profit / Loss'] / row['Close']

    # Set thresholds based on the percentage (adjust these as needed)
    if profit_loss_percentage > 0.05:  # Example: greater than 5% of close price for high profit
        return 'high profit'
    elif 0 <= profit_loss_percentage <= 0.05:  # Between 0 and 5% for low profit
        return 'low profit'
    else:
        return 'loss'

# Apply the function to categorize the values in the dataset
df_cleaned['profit_loss_category'] = df_cleaned.apply(categorize_profit_loss, axis=1)

# Check the first few rows to ensure the categorization worked
df_cleaned[['Profit / Loss', 'Close', 'profit_loss_category']]
```

```python
df_cleaned['profit_loss_category'].value_counts()
```

```python
from sklearn.model_selection import train_test_split

# Split the data into features (X) and target (y)
X = df_cleaned[['SMA-5', 'SMA-26', 'RSI', 'MACD', 'CCI', 'Close', 'Profit / Loss']]  # Adjust as needed
y = df_cleaned['profit_loss_category']

# Perform a time series split (e.g., 80% for training, 20% for testing)
train_size = int(0.8 * len(df_cleaned))
X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report

# Initialize the RandomForestClassifier with class_weight
rf_model = RandomForestClassifier(class_weight='balanced', random_state=42)

# Train the model
rf_model.fit(X_train, y_train)

# Make predictions on the test set
y_pred = rf_model.predict(X_test)

# Evaluate the model
print(classification_report(y_test, y_pred))
```

```python
from sklearn.model_selection import GridSearchCV

# Define the parameter grid to search over
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [5, 10, 20],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': ['auto', 'sqrt']
}

# Initialize the RandomForestClassifier
rf_model = RandomForestClassifier(class_weight='balanced', random_state=42)

# Set up the GridSearchCV
grid_search = GridSearchCV(estimator=rf_model, param_grid=param_grid, cv=3, n_jobs=-1, verbose=2)

# Fit the grid search model
grid_search.fit(X_train, y_train)

# Get the best parameters from the grid search
best_params = grid_search.best_params_
print("Best Parameters found: ", best_params)

# Train the model with the best parameters
best_rf_model = grid_search.best_estimator_

# Make predictions on the test set
y_pred_tuned = best_rf_model.predict(X_test)

# Evaluate the tuned Random Forest model
from sklearn.metrics import classification_report
print("Tuned Random Forest Classifier Performance:")
print(classification_report(y_test, y_pred_tuned))
```

```python
from sklearn.preprocessing import LabelEncoder
import xgboost as xgb

# Initialize and fit LabelEncoder
label_encoder = LabelEncoder()
y_train_encoded = label_encoder.fit_transform(y_train)
y_test_encoded = label_encoder.transform(y_test)

# Initialize the XGBoost model with class imbalance handling
xgb_model = xgb.XGBClassifier(objective='multi:softmax', num_class=3, scale_pos_weight=10, random_state=42)

# Fit the model
xgb_model.fit(X_train, y_train_encoded)

# Make predictions
y_pred_xgb = xgb_model.predict(X_test)

# Decode predictions back to original labels
y_pred_xgb_decoded = label_encoder.inverse_transform(y_pred_xgb)

# Evaluate the model
print("XGBoost Classifier Performance:")
print(classification_report(y_test, y_pred_xgb_decoded))

```

```python
import lightgbm as lgb
from sklearn.metrics import classification_report

# Initialize the LightGBM model with class imbalance handling
lgb_model = lgb.LGBMClassifier(class_weight='balanced', random_state=42)

# Fit the model
lgb_model.fit(X_train, y_train_encoded)

# Make predictions
y_pred_lgb = lgb_model.predict(X_test)

# Decode predictions back to original labels
y_pred_lgb_decoded = label_encoder.inverse_transform(y_pred_lgb)

# Evaluate the model
print("LightGBM Classifier Performance:")
print(classification_report(y_test, y_pred_lgb_decoded))
```

```python
from sklearn.svm import SVC
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler


# Feature Scaling (SVM generally works better with scaled features)
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Step 2: Model Initialization
svm_model = SVC(kernel='rbf', gamma='scale', C=1.0, random_state=42)

# Step 3: Training the Model
svm_model.fit(X_train, y_train)

# Step 4: Evaluation
y_pred = svm_model.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))
```

```python
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier

# Initialize the models with the best parameters from tuning

# SVM with best parameters
svm_model = SVC(C=100, gamma=0.01, kernel='rbf', probability=True)  # 'probability=True' for soft voting

# Random Forest with best parameters
rf_model = RandomForestClassifier(n_estimators=100, max_depth=10, random_state=42)

# Now we can use these in a Voting Classifier, or proceed with individual model evaluation
```

```python
from sklearn.ensemble import VotingClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Initialize the Voting Classifier with soft voting
voting_clf = VotingClassifier(
    estimators=[('svm', svm_model), ('rf', rf_model)],
    voting='soft'  # Soft voting to account for predicted probabilities
)

# Fit the Voting Classifier
voting_clf.fit(X_train, y_train)

# Make predictions on the test set
y_pred = voting_clf.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
class_report = classification_report(y_test, y_pred)

print("Ensemble Model Accuracy:", accuracy)
print("Confusion Matrix:\n", conf_matrix)
print("Classification Report:\n", class_report)
```

```python
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from sklearn.ensemble import VotingClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.model_selection import train_test_split



# Initialize XGBoost and LightGBM models
xgb_clf = XGBClassifier(random_state=42)
lgbm_clf = LGBMClassifier(random_state=42)

# Combine models in Voting Classifier (use "soft" voting if probability estimates are needed)
ensemble_clf = VotingClassifier(
    estimators=[('xgb', xgb_clf), ('lgbm', lgbm_clf)],
    voting='soft'  # Change to 'hard' if you prefer majority voting
)

# Train ensemble model
ensemble_clf.fit(X_train, y_train)

# Predict and evaluate
y_pred = ensemble_clf.predict(X_test)

# Output accuracy, confusion matrix, and classification report
print("Ensemble Model Accuracy:", accuracy_score(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))
```
